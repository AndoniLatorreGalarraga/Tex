```python
import matplotlib.pyplot as plt
plt.rcParams.update(mpl.rcParamsDefault)
import math
import random
```

Escala Lineal

```python
x, y, ey = [], [], []

a = 1
b = 0.5

k = 5

for p in range(k):
    xx = p/3 + 3/(k+2) + random.uniform(-0.1,0.1)
    x. append(xx)
    y.append(a*xx+b + random.uniform(-0.1, 0.1))
    ey.append(random.uniform(0.1, 0.3))
```

```python
ytop = [y[i]+ey[i] for i in range(len(y))]
ybot = [y[i]-ey[i] for i in range(len(y))]
```

```python
plt.axes(xlim=[0,3], ylim=[0,3],
xticks=[i/10 for i in range(0,31)],
yticks=[i/10 for i in range(0,31)],
xticklabels=['0']+['']*9+['1']+['']*9+['2']+['']*9+['3'],
yticklabels=['0']+['']*9+['1']+['']*9+['2']+['']*9+['3'])
plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
plt.scatter(x, y, marker = ".", s=100)
plt.scatter(x, ytop, marker = "_", s=100)
plt.scatter(x, ybot, marker = "_", s=100)
plt.show()
```

Escala Logaritmica

```python
x, y, ey = [], [], []

a = 3
q = 0.5

k = 5

for p in range(k):
    xx = 10**( p/3 + 3/(k+2) + random.uniform(-0.1,0.1) )
    x.append(xx)
    yy = a*xx**q + xx*random.uniform(-0.1, 0.1)
    y.append(yy)
    ey.append(yy*random.uniform(0.2, 0.5))
```

```python
ytop = [math.log(y[i],10) + math.log(math.e,10)*ey[i]/y[i] for i in range(len(y))]
ybot = [math.log(y[i],10) - math.log(math.e,10)*ey[i]/y[i] for i in range(len(y))]
```

```python
x = [math.log(i,10) for i in x]
y = [math.log(i,10) for i in y]
```

```python
plt.axes(xlim=[0,3], ylim=[0,3],
xticks=[math.log(i,10) for i in range(1,11)]+[math.log(10*i,10) for i in range(2,11)]+[math.log(100*i,10) for i in range(2,11)],
yticks=[math.log(i,10) for i in range(1,11)]+[math.log(10*i,10) for i in range(2,11)]+[math.log(100*i,10) for i in range(2,11)],
xticklabels=['$10^0$']+['']*8+['$10^1$']+['']*8+['$10^2$']+['']*8+['$10^3$'],
yticklabels=['$10^0$']+['']*8+['$10^1$']+['']*8+['$10^2$']+['']*8+['$10^3$'])
plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
plt.scatter(x, y, marker = ".", s=100)
plt.scatter(x, ytop, marker = "_", s=100)
plt.scatter(x, ybot, marker = "_", s=100)
plt.show()
```

Escala Semilogaritmica

```python
x, y, ey = [], [], []

a = 2
A = 9

k = 5

for p in range(k):
    xx = p/3 + 3/(k+2) + random.uniform(-0.1,0.1)
    x.append(xx)
    yy = A*math.exp(a*xx) + xx*random.uniform(-0.1, 0.1)
    y.append(yy)
    ey.append(yy*random.uniform(0.2, 0.5))
```

```python
ytop = [math.log(y[i],10) + math.log(math.e,10)*ey[i]/y[i] for i in range(len(y))]
ybot = [math.log(y[i],10) - math.log(math.e,10)*ey[i]/y[i] for i in range(len(y))]
```

```python
y = [math.log(i,10) for i in y]
```

```python
plt.axes(xlim=[0,3], ylim=[0,3],
xticks=[i/10 for i in range(0,31)],
yticks=[math.log(i,10) for i in range(1,11)]+[math.log(10*i,10) for i in range(2,11)]+[math.log(100*i,10) for i in range(2,11)],
xticklabels=['0']+['']*9+['1']+['']*9+['2']+['']*9+['3'],
yticklabels=['$10^0$']+['']*8+['$10^1$']+['']*8+['$10^2$']+['']*8+['$10^3$'])
plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
plt.scatter(x, y, marker = ".", s=100)
plt.scatter(x, ytop, marker = "_", s=100)
plt.scatter(x, ybot, marker = "_", s=100)
plt.show()
```
